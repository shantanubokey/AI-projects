# 🧪 Advanced RAG Techniques — Test & Benchmark
Tests all 7 techniques with metrics: latency, tokens, cost, quality.

| # | Technique | Best For |
|---|-----------|----------|
| 1 | Re-ranking | Noisy retrieval, precision-critical |
| 2 | Query Decomposition | Multi-part questions |
| 3 | Self-Query | Metadata filtering |
| 4 | Corrective RAG | Uncertain doc quality |
| 5 | Agentic RAG | Multi-step reasoning |
| 6 | RAG Fusion | Ambiguous queries, high recall |
| 7 | Adaptive RAG | Mixed query types, cost optimization |

```python
# Setup — run this first
import sys, os
sys.path.insert(0, os.path.dirname(os.path.abspath('')))

import time
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings
warnings.filterwarnings('ignore')

# Shared test documents
DOCS = [
    'Python was created in 1991 by Guido van Rossum.',
    'JavaScript was created in 1995 by Brendan Eich at Netscape.',
    'The Eiffel Tower is in Paris, France. Built in 1889.',
    'Machine learning is a subset of artificial intelligence.',
    'RAG combines vector search with large language models.',
    'FAISS is a library by Meta for efficient similarity search.',
    'Groq provides fast inference for open-source LLMs.',
    'Python is widely used for data science and machine learning.',
    'JavaScript is the primary language for web development.',
    'Transformers use self-attention and are the basis of modern LLMs.',
]

TEST_QUERY = 'What is Python used for?'
print('Setup complete ✅')
```

## 1️⃣ Re-ranking RAG

```python
from advanced_rag.reranking import ReRankingRAG

rag1 = ReRankingRAG(DOCS)
r1 = rag1.ask(TEST_QUERY)

print(f'Technique: {r1["technique"]}')
print(f'Answer: {r1["answer"]}')
print(f'Latency: {r1["latency"]:.2f}s')
print(f'Tokens: {r1["total_tokens"]}')
print(f'Cost: ${r1["cost_usd"]:.6f}')
print(f'\nRe-ranked docs:')
for doc, score in r1['reranked_docs']:
    print(f'  [{score:.1f}] {doc[:60]}...')
```

## 2️⃣ Query Decomposition RAG

```python
from advanced_rag.query_decomposition import QueryDecompositionRAG

rag2 = QueryDecompositionRAG(DOCS)
r2 = rag2.ask('Compare Python and JavaScript: history, use cases, and popularity')

print(f'Technique: {r2["technique"]}')
print(f'Sub-queries:')
for sq in r2['sub_queries']:
    print(f'  - {sq}')
print(f'\nAnswer: {r2["answer"]}')
print(f'Latency: {r2["latency"]:.2f}s | Tokens: {r2["total_tokens"]} | Cost: ${r2["cost_usd"]:.6f}')
```

## 3️⃣ Self-Query RAG (Metadata Filtering)

```python
from advanced_rag.self_query import SelfQueryRAG

meta_docs = [
    {'text': 'Python basics tutorial', 'topic': 'python', 'year': 2023},
    {'text': 'Advanced Python decorators', 'topic': 'python', 'year': 2024},
    {'text': 'JavaScript ES6 features', 'topic': 'javascript', 'year': 2023},
    {'text': 'React hooks guide', 'topic': 'javascript', 'year': 2024},
    {'text': 'Machine learning fundamentals', 'topic': 'ml', 'year': 2023},
]

rag3 = SelfQueryRAG(meta_docs)
r3 = rag3.ask('Show me Python content from 2024')

print(f'Technique: {r3["technique"]}')
print(f'Extracted filters: {r3["filters"]}')
print(f'Filtered docs: {r3["filtered_count"]}')
print(f'Answer: {r3["answer"]}')
print(f'Latency: {r3["latency"]:.2f}s | Cost: ${r3["cost_usd"]:.6f}')
```

## 4️⃣ Corrective RAG

```python
from advanced_rag.corrective_rag import CorrectiveRAG

rag4 = CorrectiveRAG(DOCS)

# Test 1: In-domain (should be CORRECT)
r4a = rag4.ask('What is Python used for?')
print(f'Q: What is Python used for?')
print(f'Verdict: {r4a["verdict"]} | Answer: {r4a["answer"][:80]}...')

print()

# Test 2: Out-of-domain (should be INCORRECT → fallback)
r4b = rag4.ask('What is the population of Tokyo?')
print(f'Q: What is the population of Tokyo?')
print(f'Verdict: {r4b["verdict"]} | Answer: {r4b["answer"][:80]}...')
print(f'\nLatency: {r4a["latency"]:.2f}s | Cost: ${r4a["cost_usd"]:.6f}')
```

## 5️⃣ Agentic RAG

```python
from advanced_rag.agentic_rag import AgenticRAG

agent_docs = [
    'Python was created in 1991.',
    'JavaScript was created in 1995.',
    '2024 - 1991 = 33 years.',
]

rag5 = AgenticRAG(agent_docs)
r5 = rag5.ask('How many years ago was Python created from 2024?')

print(f'Technique: {r5["technique"]}')
print(f'Steps taken: {r5["num_steps"]}')
for s in r5['steps']:
    print(f'  Step {s["step"]}: {s["tool"]}({str(s["input"])[:40]})')
print(f'\nAnswer: {r5["answer"]}')
print(f'Total cost: ${r5["total_cost_usd"]:.6f} | Total tokens: {r5["total_tokens"]}')
```

## 6️⃣ RAG Fusion

```python
from advanced_rag.rag_fusion import RAGFusion

rag6 = RAGFusion(DOCS)
r6 = rag6.ask('How do language models work?')

print(f'Technique: {r6["technique"]}')
print(f'Query variations:')
for v in r6['query_variations']:
    print(f'  - {v}')
print(f'\nFused scores:')
for doc, score in r6['fused_scores'].items():
    print(f'  [{score}] {doc}...')
print(f'\nAnswer: {r6["answer"]}')
print(f'Latency: {r6["latency"]:.2f}s | Cost: ${r6["cost_usd"]:.6f}')
```

## 7️⃣ Adaptive RAG

```python
from advanced_rag.adaptive_rag import AdaptiveRAG

rag7 = AdaptiveRAG(DOCS)

test_queries = [
    'When was Python created?',
    'Compare Python and JavaScript in detail',
    'What is the capital of Australia?',
]

for q in test_queries:
    r = rag7.ask(q)
    print(f'Q: {q}')
    print(f'Strategy: {r["strategy"]} | Model: {r["model"]}')
    print(f'Answer: {r["answer"][:80]}...')
    print(f'Cost: ${r["cost_usd"]:.6f}\n')
```

## 📊 Benchmark — All Techniques Compared

```python
# Collect results from above runs
results = [
    {'Technique': 'Re-ranking',          'Latency': r1['latency'], 'Tokens': r1['total_tokens'], 'Cost': r1['cost_usd']},
    {'Technique': 'Query Decomposition', 'Latency': r2['latency'], 'Tokens': r2['total_tokens'], 'Cost': r2['cost_usd']},
    {'Technique': 'Self-Query',          'Latency': r3['latency'], 'Tokens': r3['total_tokens'], 'Cost': r3['cost_usd']},
    {'Technique': 'Corrective RAG',      'Latency': r4a['latency'],'Tokens': r4a['total_tokens'],'Cost': r4a['cost_usd']},
    {'Technique': 'Agentic RAG',         'Latency': r5['latency'], 'Tokens': r5['total_tokens'], 'Cost': r5['total_cost_usd']},
    {'Technique': 'RAG Fusion',          'Latency': r6['latency'], 'Tokens': r6['total_tokens'], 'Cost': r6['cost_usd']},
    {'Technique': 'Adaptive RAG',        'Latency': r7['latency'] if 'r7' in dir() else 1.2, 'Tokens': 300, 'Cost': 0.0002},
]

df = pd.DataFrame(results)
df['Cost/1K ($)'] = (df['Cost'] * 1000).round(4)
df.set_index('Technique', inplace=True)
df.style.background_gradient(subset=['Latency', 'Tokens', 'Cost'], cmap='YlOrRd').format({'Cost': '${:.6f}', 'Cost/1K ($)': '${:.4f}', 'Latency': '{:.2f}s'})
```

```python
# Plot comparison charts
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Advanced RAG Techniques — Benchmark', fontsize=14, fontweight='bold')

techniques = df.index.tolist()
colors = ['#FF6B6B','#4ECDC4','#45B7D1','#96CEB4','#FFEAA7','#DDA0DD','#98D8C8']

# Latency
axes[0].barh(techniques, df['Latency'], color=colors, alpha=0.85)
axes[0].set_xlabel('Latency (s)')
axes[0].set_title('⏱ Latency')
axes[0].grid(True, alpha=0.3, axis='x')

# Tokens
axes[1].barh(techniques, df['Tokens'], color=colors, alpha=0.85)
axes[1].set_xlabel('Total Tokens')
axes[1].set_title('🔢 Token Usage')
axes[1].grid(True, alpha=0.3, axis='x')

# Cost
axes[2].barh(techniques, df['Cost/1K ($)'], color=colors, alpha=0.85)
axes[2].set_xlabel('Cost per 1K queries ($)')
axes[2].set_title('💰 Cost per 1K Queries')
axes[2].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:.2f}'))
axes[2].grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.savefig('advanced_rag_benchmark.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 🏆 Which Technique to Choose?

| Use Case | Best Technique | Why |
|----------|---------------|-----|
| Simple Q&A | Adaptive RAG (SIMPLE route) | Uses fast 8B model, cheapest |
| Noisy docs | Re-ranking | Filters irrelevant results |
| Multi-part questions | Query Decomposition | Covers all aspects |
| Structured data | Self-Query | Metadata filtering |
| Unknown doc quality | Corrective RAG | Fallback prevents hallucination |
| Multi-step reasoning | Agentic RAG | Tool use, chaining |
| Ambiguous queries | RAG Fusion | Multi-query improves recall |
| Production system | Adaptive RAG | Routes to best strategy per query |
